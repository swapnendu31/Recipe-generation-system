```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the GEFX file
G = nx.read_gexf("bipartite_recipe.gexf")

# Check if the graph is bipartite
from networkx.algorithms import bipartite
is_bipartite = bipartite.is_bipartite(G)

# If bipartite, get the two sets
if is_bipartite:
    set1, set2 = bipartite.sets(G)
    pos = {node: (0, index) for index, node in enumerate(set1)}
    pos.update({node: (1, index) for index, node in enumerate(set2)})
else:
    pos = nx.spring_layout(G)  # Fall back to a generic layout
pos = nx.drawing.layout.bipartite_layout(G, set1)
# Draw the graph
nx.draw(G, pos, with_labels=True, node_color="skyblue", edge_color="gray")

```

```python
from semantic_kernel import Kernel
from service_settings import ServiceSettings
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion
from semantic_kernel.functions import KernelArguments

async def return_desc(re: str):
    kernel = Kernel()
    service_settings = ServiceSettings.create()
    kernel.remove_all_services()
    service_id = "default"
    kernel.add_service(
        AzureChatCompletion(
            service_id=service_id,
        ),
    )
    plugin = kernel.add_plugin(parent_directory="", plugin_name="FlugIn")
    desc_function = plugin["desc"]
    description = await kernel.invoke(desc_function, KernelArguments(input=re))
    return description
    
async def get_description(item:str):
    recipe_text = item  # Add the recipe text you want a description for
    description = await return_desc(recipe_text)
    return description
```

```python
import re

def recipe_dict(li:str,item:str):
    li = li[0]
    re_dict = {}
    item = '-'.join(item.split(','))
    li_re = []
    for i in li:
        dt = {}
        name = i.replace('Recipe','')
        name = name.replace('-','')
        cleaned_name = re.sub(r"\(.*?\)", "", name).strip()
        dt['name'] = name
        dt['description'] = "des"
        dt['instruction'] = "sa"
        dt['tags'] = ''.join(df[df['RecipeName']==i]['Vegetables']).split(',')
        dt['calories']=list(df[df['RecipeName']==i]['TotalTimeInMins'])[0]
        dt['cuisine'] = list(df[df['RecipeName']==i]['Cuisine'])[0]
        li_re.append(dt)
    re_dict[item]=li_re
    return re_dict
```
